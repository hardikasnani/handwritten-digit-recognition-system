```python
#importing relevant libraries
import torch
import torch.nn as nn
import torch.nn.functional as Func
from torch.autograd import Variable
import torch.optim as optim
import torch.utils.data as data
import random
from scipy.io import savemat
import os
from os import path
from sklearn.preprocessing import normalize
from torch.nn.utils import clip_grad_norm
import torch.nn.parallel.data_parallel as data_parallel
from sklearn.metrics import confusion_matrix
```

```python
#defining important hyperparameters
class hyperparam:
  number_hid3_units = 30
  in_channel = 1
  number_classes = 10
  input_dimension = 28*28
  learning_rate = 0.05
  number_epochs = 50
  mini_batch_size = 50
```

```python
#using DataLoader to perform mini batch learning
class my_dataset(data.Dataset):
  def __init__(self, input_data, labels):
    self.features = input_data
    self.labels = labels

  def __len__(self):
    return len(self.labels)

  def __getitem__(self, index):
    indexed_feature = self.features.iloc[index,:]
    indexed_label = self.labels.iloc[index]
    indexed_feature = indexed_feature.values
    indexed_label = indexed_label.values

    return indexed_feature,indexed_label
```

```python
import pandas as pd

#creating DataLoader instances for all the datasets
params = {'batch_size' : hyperparam.mini_batch_size,
          'shuffle' : True,
          'num_workers' : 6,
          'drop_last' : False,
          'pin_memory' : True
          }

# uncomment this for final submission
y_train_strat = pd.DataFrame({'digit': y_train_strat})
y_val_strat = pd.DataFrame({'digit': y_val_strat})
y_test = pd.DataFrame({'digit': y_test})

training_set = my_dataset(X_train_strat, y_train_strat)
validation_set = my_dataset(X_val_strat, y_val_strat)
testing_set = my_dataset(X_test, y_test)

training_gen = data.DataLoader(training_set, **params)
validation_gen = data.DataLoader(validation_set, **params)
testing_gen = data.DataLoader(testing_set, **params)
```

```python
torch.cuda.is_available()
def get_default_device():
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
device = get_default_device()
device
```

```python
import numpy as np

class early_stopping:
    def __init__(self, patience=7, delta=0, path='checkpoint.pt', trace_func=print):
        self.patience = patience
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf
        self.delta = delta
        self.path = path
        self.trace_func = trace_func

    def check_for_early_stopping(self, val_loss, model):

        score = -val_loss

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            self.trace_func(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        torch.save(model.state_dict(), self.path)
        self.val_loss_min = val_loss
```

```python
#defining the neural network
class CNN(nn.Module):
  def __init__(self):
    super(CNN,self).__init__()

    self.n_hid3_units = hyperparam.number_hid3_units
    self.n_out_units = hyperparam.number_classes
    self.inp_dimension = hyperparam.input_dimension

    self.conv1 = nn.Conv2d(in_channels=hyperparam.in_channel, out_channels=32, kernel_size=(3,3), stride=(1,1))
    self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3,3), stride=(1,1))
    self.fc1 = nn.Linear(1600, self.n_hid3_units)
    self.fc2 = nn.Linear(self.n_hid3_units, self.n_out_units)

    #initial values for the network
    nn.init.xavier_uniform_(self.conv1.weight)
    self.conv1.bias.data.fill_(0.0)

    nn.init.xavier_uniform_(self.conv2.weight)
    self.conv2.bias.data.fill_(0.0)

    nn.init.xavier_uniform_(self.fc1.weight)
    self.fc1.bias.data.fill_(0.0)

    nn.init.xavier_uniform_(self.fc2.weight)
    self.fc2.bias.data.fill_(0.0)
  
  #defining the forward pass of the CNN with relu activation function
  def forward(self,sig):
    sig = self.conv1(sig)
    sig = Func.max_pool2d(sig, kernel_size=(2,2), stride=(2,2))
    sig = Func.relu(sig)

    sig = self.conv2(sig)
    sig = Func.max_pool2d(sig, kernel_size=(2,2), stride=(2,2))
    sig = Func.relu(sig)

    sig = sig.reshape(sig.shape[0], -1)

    sig = self.fc1(sig)
    sig = Func.relu(sig)

    sig = self.fc2(sig)

    return sig
```

## **Using RMSprop optimizer**

```python
cnn_model = CNN()
if torch.cuda.is_available():
    cnn_model.cuda()
optimizer = torch.optim.RMSprop(cnn_model.parameters(), lr=0.01, momentum=0)
loss = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.9)
```

```python
#training the cnn
training_avgLoss_list = []
training_accuracy_list = []
validation_avgLoss_list = []
validation_accuracy_list = []

reg_early_stopping = early_stopping(path='checkpoint_RMSprop.pt')

#looping over epochs
for epoch in range(hyperparam.number_epochs):
  training_num_correct = 0
  training_num_samples = 0
  training_total_loss = 0.0

  validation_num_correct = 0
  validation_num_samples = 0
  validation_total_loss = 0.0

  #training
  cnn_model.train(True)

  with torch.set_grad_enabled(True):
    for local_batch, local_labels in training_gen:
      optimizer.zero_grad()

      local_batch = local_batch.view(hyperparam.mini_batch_size,-1,28,28)

      local_batch = local_batch.to(device).float()
      local_labels = local_labels.to(device).float()
      local_batch, local_labels = Variable(local_batch), Variable(local_labels)

      #model computations
      out1 = cnn_model(local_batch)

      #cross entropy loss calculation
      pLoss = loss(out1,torch.flatten(local_labels.long()))
      #pLoss = loss(out1, local_labels.long())
      training_total_loss += pLoss * hyperparam.mini_batch_size

      #backpropagation
      #gradient calculation
      pLoss.backward()

      #weight update
      optimizer.step()

      sel_class = torch.argmax(out1,dim=1)

      training_num_correct += sel_class.eq(torch.flatten(local_labels.long())).sum().item()
      training_num_samples += hyperparam.mini_batch_size

    training_avgLoss = training_total_loss/len(training_gen.dataset)
    training_avgLoss_list.append(training_avgLoss)

    training_accuracy = training_num_correct/training_num_samples
    training_accuracy_list.append(training_accuracy)

  #validation
  with torch.set_grad_enabled(False):
    cnn_model.eval()

    for local_batch, local_labels in validation_gen:

      local_batch = local_batch.view(hyperparam.mini_batch_size,-1,28,28)

      local_batch = local_batch.to(device).float()
      local_labels = local_labels.to(device).float()
      local_batch, local_labels = Variable(local_batch), Variable(local_labels)

      #model computations
      out1 = cnn_model(local_batch)

      #CrossEntropy loss calculation
      ploss = loss(out1,torch.flatten(local_labels.long()))
      validation_total_loss += ploss * hyperparam.mini_batch_size

      sel_class = torch.argmax(out1, dim=1)

      validation_num_correct += sel_class.eq(torch.flatten(local_labels.long())).sum().item()
      validation_num_samples += hyperparam.mini_batch_size
    
    validation_avgLoss = validation_total_loss / len(validation_gen.dataset)
    validation_avgLoss_list.append(validation_avgLoss)

    reg_early_stopping.check_for_early_stopping(validation_avgLoss, cnn_model)

    if reg_early_stopping.early_stop:
      print("Early Stopping")
      break      

    validation_accuracy = validation_num_correct / validation_num_samples
    validation_accuracy_list.append(validation_accuracy)
    scheduler.step()
    print('Epoch-{0} lr: {1}'.format(epoch, optimizer.param_groups[0]['lr']))
    if epoch % 10 == 0:
      print()
    print(f'training results -> Epoch: {epoch} Loss: {training_avgLoss} Accuracy: {training_accuracy}')
    print(f'validation results -> Epoch: {epoch} Loss: {validation_avgLoss} Accuracy: {validation_accuracy}')
```

```python
#training the cnn
training_avgLoss_list = []
training_accuracy_list = []
validation_avgLoss_list = []
validation_accuracy_list = []

reg_early_stopping = early_stopping(path='checkpoint_nesterov.pt')

#looping over epochs
for epoch in range(hyperparam.number_epochs):
  training_num_correct = 0
  training_num_samples = 0
  training_total_loss = 0.0

  validation_num_correct = 0
  validation_num_samples = 0
  validation_total_loss = 0.0

  #training
  cnn_model.train(True)

  with torch.set_grad_enabled(True):
    for local_batch, local_labels in training_gen:
      optimizer.zero_grad()

      local_batch = local_batch.view(hyperparam.mini_batch_size,-1,28,28)

      local_batch = local_batch.to(device).float()
      local_labels = local_labels.to(device).float()
      local_batch, local_labels = Variable(local_batch), Variable(local_labels)

      #model computations
      out1 = cnn_model(local_batch)

      #cross entropy loss calculation
      pLoss = loss(out1,torch.flatten(local_labels.long()))
      #pLoss = loss(out1, local_labels.long())
      training_total_loss += pLoss * hyperparam.mini_batch_size

      #backpropagation
      #gradient calculation
      pLoss.backward()

      #weight update
      optimizer.step()

      sel_class = torch.argmax(out1,dim=1)

      training_num_correct += sel_class.eq(torch.flatten(local_labels.long())).sum().item()
      training_num_samples += hyperparam.mini_batch_size

    training_avgLoss = training_total_loss/len(training_gen.dataset)
    training_avgLoss_list.append(training_avgLoss)

    training_accuracy = training_num_correct/training_num_samples
    training_accuracy_list.append(training_accuracy)

  #validation
  with torch.set_grad_enabled(False):
    cnn_model.eval()

    for local_batch, local_labels in validation_gen:

      local_batch = local_batch.view(hyperparam.mini_batch_size,-1,28,28)

      local_batch = local_batch.to(device).float()
      local_labels = local_labels.to(device).float()
      local_batch, local_labels = Variable(local_batch), Variable(local_labels)

      #model computations
      out1 = cnn_model(local_batch)

      #CrossEntropy loss calculation
      ploss = loss(out1,torch.flatten(local_labels.long()))
      validation_total_loss += ploss * hyperparam.mini_batch_size

      sel_class = torch.argmax(out1, dim=1)

      validation_num_correct += sel_class.eq(torch.flatten(local_labels.long())).sum().item()
      validation_num_samples += hyperparam.mini_batch_size
    
    validation_avgLoss = validation_total_loss / len(validation_gen.dataset)
    validation_avgLoss_list.append(validation_avgLoss)

    reg_early_stopping.check_for_early_stopping(validation_avgLoss, cnn_model)

    if reg_early_stopping.early_stop:
      print("Early Stopping")
      break      

    validation_accuracy = validation_num_correct / validation_num_samples
    validation_accuracy_list.append(validation_accuracy)
    scheduler.step()
    print('Epoch-{0} lr: {1}'.format(epoch, optimizer.param_groups[0]['lr']))
    if epoch % 10 == 0:
      print()
    print(f'training results -> Epoch: {epoch} Loss: {training_avgLoss} Accuracy: {training_accuracy}')
    print(f'validation results -> Epoch: {epoch} Loss: {validation_avgLoss} Accuracy: {validation_accuracy}')
```
